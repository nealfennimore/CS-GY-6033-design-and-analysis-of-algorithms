## Big O

The upper bound asymptotic growth rate

$f(n) \text{ is } O(g(n))$

---

Definition: $ f(n) \leq C \times g(n) $
- $C$: positive constant
- True for $ n \geq k $

### Job

- Create an upper bound
- Identify the constant $C$
- State which values of $n$ are valid

### Example

$$
\begin{align}
\text{Show } f(n) = n^2 + n + 3 \text{ is } O(n^2) \newline
\text{Goal: } f(n) \leq C \times n^2 \newline
n^2 + n + 3 \leq n^2 + n^2 + 3n^2 \newline
n^2 + n + 3 \leq 5n^2 \newline
\therefore f(n) \leq 5n^2 \text{ for } n \geq 1 \newline
f(n) \text{ is } O(n^2)
\end{align}
$$

```python
t1 = Opt(fn=lambda n: n ** 2 + n + 3, label="n^2 + n + 3")
t2 = Opt(lambda n: 5 * n ** 2, "5n^2")
t3 = Opt(lambda n: n, "n")

Plotter.render_plot(100, t1, t2, t3)
```

## Big Omega

The lower bound asymptotic growth rate

$f(n)\ is\ \Omega(g(n))$

---

Definition: $ f(n) \geq C \times g(n) $
- $C$: positive constant
- True for $ n \geq k $

### Job

- Create a lower bound
- Identify the constant _C_
- State which values of _n_ are valid

### Example

$$
\begin{align}
\text{Show } f(n) = n^2 + \sqrt(n) \log(n) + 3 \text{ is } \Omega(n^2) \newline
\text{Goal: } f(n) \geq C \times n^2 \newline
n^2 + \sqrt(n) \log(n) + 3 \geq n^2 \newline
\therefore f(n) \geq 1 \times n^2 \newline
f(n) \text{ is } \Omega(n^2)
\end{align}
$$

```python
import math

t1 = Opt(lambda n: n ** 2 + math.sqrt(n) * math.log(n) + 3, "n^2 + sqrt(n) * log(n) + 3")
t2 = Opt(lambda n: n ** 2, "n^2")

Plotter.render_plot(25, t1, t2)
```

## Big Theta

Has upper $ (O) $ and lower $ (\Omega) $ bounds

$f(n)\text{ is }\Theta(g(n))$

---

Definition: $f(n)$ is both:
- $O(g(n))$
- $\Omega(g(n))$

### Example

$$
\text{Show } f(n) = 3n^2 + \log(n) + 2^n \text{ is } \Theta(2^n)
$$

#### Big O
$$
\begin{align}
O(2^n) \text{ show } f(n) \leq C \times 2^n \newline
3n^2 + \log(n) + 2^n \leq 3 \times 2^n + 2^n + 2^n\newline
3n^2 + \log(n) + 2^n \leq 5 \times 2^n \newline
\therefore f(n) \text{ is } O(2^n)
\end{align}
$$

#### Big Omega
$$
\begin{align}
\Omega(2^n) \text{ show } f(n) \geq C \times 2^n \newline
3n^2 + \log(n) + 2^n \geq 2^n \newline
\therefore f(n) \geq 1 \times 2^n \text{ for } n \geq 1 \newline
\therefore f(n) \text{ is } \Omega(2^n)
\end{align}
$$

#### Answer
$$
f(n) \text{ is } \Theta(2^n)
$$

```python
t1 = Opt(lambda n: 5 * (2 ** n), "O(2^n) = 5 * 2^n")
t2 = Opt(lambda n: 2 ** n, "Theta(2^n) = 2^n")
t3 = Opt(lambda n: 2 ** n, "Omega(2^n) = 2^n")

Plotter.render_plot(100, t1, t2, t3)
```
